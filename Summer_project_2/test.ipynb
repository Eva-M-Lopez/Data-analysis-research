```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


df = pd.read_csv("titanic_data.csv")
#reader
```

```python
# Filter the data to show only those who survived
survivors = df[df['Survived'] == 1]

# Group the data by class and count the number of survivors in each class
survivors_by_class = survivors['Pclass'].value_counts().sort_index()

# Plot the results using a bar graph
plt.bar(survivors_by_class.index, survivors_by_class.values)

# Add labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')
plt.title('Number of Survivors in Each Class')

# Show the plot
plt.show()
```

```python
# Filter the data to show only those who survived
survivors = df[df['Survived'] == 1]

# Group the data by gender and count the number of survivors in each class
survivors_by_gender = survivors['Sex'].value_counts().sort_index()

# Plot the results using a bar graph
plt.bar(survivors_by_gender.index, survivors_by_gender.values)

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Number of Survivors')
plt.title('Number of Survivors in Each Sex')

# Show the plot
plt.show()
```

```python
# Filter the data to show only those who survived
survivors = df[df['Survived'] == 1]

# Group the data by gender and count the number of survivors in each class
survivors_by_fare = survivors['Fare'].value_counts().sort_index()

# Plot the results using a bar graph
plt.bar(survivors_by_fare.index, survivors_by_fare.values)

# Add labels and title
plt.xlabel('Fare')
plt.ylabel('Number of Survivors')
plt.title('Number of Survivors in Each Fare')

# Show the plot
plt.show()
```

```python
# Filter the data to show only those who survived
Fare = df[df['Fare'] == 1]

# Group the data by gender and count the number of survivors in each class
survivors_by_Pclass = Fare['Pclass'].value_counts().sort_index()

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Pclass'], df['Fare'], alpha=0.5)

# Add labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Scatter Plot of Passenger Class vs Fare')

# Show the plot
plt.show()
```

```python
# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define data for the scatter plot
x = df['Fare']    # Fares
y = df['Survived']  # Survival (0 or 1)
z = df['Pclass']   # Passenger class

# Create a 3D scatter plot
scatter = ax.scatter(x, y, z, c=z, cmap='viridis')

# Add labels and title
ax.set_xlabel('Fare')
ax.set_ylabel('Survived')
ax.set_zlabel('Passenger Class')
ax.set_title('3D Scatter Plot of Fare, Survival, and Passenger Class')

# Add a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Passenger Class')

# Show the plot
plt.show()
```
